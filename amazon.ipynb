```python
high_rating_count = 0

for rating in data["rating"]:
    if pd.notna(rating) and rating >= 4:
        high_rating_count += 1

print("Products with rating >= 4:", high_rating_count)
```

```python
data["discount_percentage"].dtype
```

```python
data["discount_percentage"] = (
    data["discount_percentage"]
    .astype(str)
    .str.replace("%", "", regex=False)
)
```

```python
import pandas as pd

data["discount_percentage"] = pd.to_numeric(
    data["discount_percentage"], errors="coerce"
)
```

```python
data["discount_percentage"].dtype
```

```python
def discount_type(discount):
    if discount >= 50:
        return "High Discount"
    elif discount >= 20:
        return "Medium Discount"
    else:
        return "Low Discount"

data["discount_type"] = data["discount_percentage"].apply(discount_type)
data[["product_name", "discount_percentage", "discount_type"]].head()
```

```python
def average_rating(df):
    return df["rating"].mean()

print("Average Product Rating:", average_rating(data))
```
